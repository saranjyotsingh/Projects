```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import os
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
train_df.info()
test_df.info()
```

```python
train_df.describe(include='all') 
# 'O' just objests
# 'all' both nuberical and objests
# default behaviour is only numbers
```

```python
train_df[["Sex", "Survived"]].groupby(['Sex'], as_index=False).mean().sort_values(by='Survived', ascending=False)
train_df=train_df.drop(['Ticket','Cabin'], axis=1)
print(len(train_df))
train_df = train_df.dropna()
print(len(train_df))
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
predictors_train=train_df
targ_train=predictors_train.pop("Survived")
print(len(predictors_train))
print(len(targ_train))
nVal=list(predictors_train.dtypes[predictors_train.dtypes!="object"].index)
print(nVal)
#model.fit(predictors_train, targ_train)
```
